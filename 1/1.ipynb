# Projétil com resistência do ar

A modelagem matemática de sistemas dinâmicos é o pilar fundamental para a compreensão e previsão de fenômenos físicos no mundo real. Este projeto, intitulado "Projétil com resistência do ar", tem como objetivo desenvolver uma modelagem determinística e estocástica da trajetória de uma massa pontual submetida simultaneamente a um campo gravitacional uniforme e a forças de arrasto aerodinâmico. Ao abandonar a aproximação parabólica ideal, o sistema de equações diferenciais resultante torna-se não linear e acoplado, impossibilitando, na maioria dos casos práticos, a obtenção de soluções analíticas fechadas em termos de funções elementares.

Consequentemente, este problema exige a aplicação de Métodos Computacionais para a resolução do sistema dinâmico, e de Métodos Estatísticos para avaliar como a variabilidade dos parâmetros iniciais — flutuações na densidade do ar, incertezas no coeficiente de arrasto e na velocidade de lançamento — afeta o espaço de fase e o ponto de impacto do projétil. O presente estudo documenta a formulação do problema, a implementação algorítmica da solução e a análise estatística dos resultados obtidos.


---
#### Problema:
Considere um projétil modelado como uma partícula de massa $m$ que é lançado do solo, a partir da origem, com velocidade $v_0$ e ângulo $\theta$ acima da horizontal. O projétil move-se no interior de um fluido (o ar) nas proximidades da superfície terrestre. Durante o voo, o projétil está submetido a duas forças fundamentais:

- **Força Gravitacional**: Assumindo que a variação de altitude é desprezível em relação ao raio da Terra, o campo gravitacional $g$ é considerado uniforme. Se o eixo $y$ for orientado positivamente para cima, a força peso é $\vec{P}=-mg\hat{j}$.

- **Força de Arrasto Aerodinâmico**: O fluido exerce uma força dissipativa $F_d$ que atua no sentido oposto ao vetor velocidade instantânea $v=\dot{r}$.

Queremos responder, quantitativamente:

1. *Quanto o arrasto reduz o alcance em relação à idealização sem ar?*
2. *Como a trajetória deixa de ser simétrica?*
3. *O ângulo de alcance máximo ainda é 45°? Se não, quanto vale?*


O obstáculo é que, no regime de arrasto que se aplica a um projétil real (arrasto quadrático na velocidade), as equações de movimento não têm solução analítica fechada. É exatamente esse o ponto de entrada dos métodos computacionais.
